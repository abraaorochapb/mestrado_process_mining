```python
import pandas as pd
import numpy as np
import pm4py as pm4
```

```python
dataframe = pd.read_csv('/home/abraao/Downloads/running-example.csv', sep=';')
print(dataframe.columns.tolist())
dataframe.head()
```

```python
dataframe = dataframe.rename(columns={'case_id': 'case:case_id'})

dataframe = pm4.format_dataframe(
    dataframe,
    case_id='case:case_id',
    activity_key='activity',
    timestamp_key='timestamp'
)
```

```python
event_log = pm4.convert_to_event_log(dataframe)

print(event_log)
```

```python
dfg, sa, ea = pm4.discover_directly_follows_graph(event_log)
pm4.view_dfg(dfg, sa, ea)
```
